```python
import numpy as np
import matplotlib.pyplot as plt

# example data
x = np.array([8,10,12])
y = np.array([10,13,16])


x_mean = np.mean(x)
y_mean = np.mean(y)

# calculate m
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
m = numerator / denominator

# calculate b
b = y_mean - m * x_mean


print(f"b (intercept): {b}")
print(f"m (slope): {m}")

# visualization
plt.scatter(x, y, color='blue')
plt.plot(x, y, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression with m and b')
plt.show()
```
